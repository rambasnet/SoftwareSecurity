## Conditional Statements

```python
if True:
    print("this line...")
    print(1)
```

```python
if name == "John Smith":
    print("You're John Smith!")
else:
    print("You're not John Smith!")
```

```python
num = int(input('Enter an integer:' ))
if num == 0:
    print(f'{num} is Zero.')
elif num%2 == 0:
    print(f'{num} is Even.')
else:
    print(f'{num} is odd.')
```

## Loops

```python
for i in range(10):
    print(f'{i+1} Mississippi!')
```

```python
while True:
    num = int(input('Enter a number: '))
    if num == 0:
        break
    else:
        print(f'You entered {num}')
```

## System command
- execute system/shell commands from with Python

### os.system()

```python
import os
```

```python
cmd = "git --version"
ret_val = os.system(cmd)
print(f'returned: {ret_val}')
```

### subprocess.call()
- subprocess.check_output() allows to store the output in a variable

```python
import subprocess
```

```python
ret_val = subprocess.call(cmd, shell=True)
```

```python
ret_val
```

```python
output = subprocess.check_output(cmd, shell=True)
```

```python
output
```

```python
# print the output binary in ASCII
print(*output)
```

```python
# print the returned binary as a utf-8 encoded string
print(output.decode('utf-8'))
```
